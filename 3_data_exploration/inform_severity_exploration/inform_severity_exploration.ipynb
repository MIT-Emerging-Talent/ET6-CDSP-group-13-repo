```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the prepared dataset
file_path = r'D:\ET6-CDSP-group-13-repo\2_data_preparation\processed\crisis_severity_2019-24.csv'
df = pd.read_csv(file_path)

# 1. Overall severity score distribution
plt.figure(figsize=(7, 4))
sns.histplot(df['severity_score'].dropna(), bins=20, kde=True)
plt.title('Distribution of Severity Score (All Country-Years)')
plt.xlabel('Severity Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 2. Severity score by year
plt.figure(figsize=(8, 5))
sns.boxplot(x='Year', y='severity_score', data=df)
plt.title('Severity Score by Year')
plt.tight_layout()
plt.show()

# 3. Severity score by country (top 10 countries with most entries)
top_countries = df['ISO3'].value_counts().head(10).index
plt.figure(figsize=(10, 6))
sns.boxplot(x='ISO3', y='severity_score', data=df[df['ISO3'].isin(top_countries)])
plt.title('Severity Score by Country (Top 10)')
plt.tight_layout()
plt.show()

# 4. Average severity by country (all countries)
avg_severity_country = df.groupby('ISO3')['severity_score'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 5))
avg_severity_country.plot(kind='bar')
plt.title('Average Severity Score by Country')
plt.xlabel('Country (ISO3)')
plt.ylabel('Average Severity Score')
plt.tight_layout()
plt.show()

# 5. Average severity by year
avg_severity_year = df.groupby('Year')['severity_score'].mean()
plt.figure(figsize=(8, 4))
avg_severity_year.plot(marker='o')
plt.title('Average Severity Score by Year')
plt.xlabel('Year')
plt.ylabel('Average Severity Score')
plt.tight_layout()
plt.show()
```
